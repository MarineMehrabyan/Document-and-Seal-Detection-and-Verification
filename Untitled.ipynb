```python
import os
import matplotlib.pyplot as plt
from PIL import Image

def plot_images_from_directory(directory):
    # Get a list of files in the directory
    files = os.listdir(directory)
    image_files = [f for f in files if f.endswith(('.png', '.jpg', '.jpeg', '.PNG'))]

    # Check how many images are available
    num_images_to_plot = min(5, len(image_files))  # Only display up to 5 images

    # Create a figure with a single row and multiple columns
    fig, axes = plt.subplots(1, num_images_to_plot, figsize=(20, 4))  # Set figsize appropriately
    for i in range(num_images_to_plot):
        img_file = image_files[i]
        img_path = os.path.join(directory, img_file)
        image = Image.open(img_path)

        if num_images_to_plot == 1:  # If there's only one image to display
            ax = axes
        else:
            ax = axes[i]

        ax.imshow(image)
        ax.axis('off')  # Turn off axis numbers and ticks

    plt.tight_layout()  # Adjust layout to prevent overlapping
    plt.show()

# Directory containing images
image_directory = 'signature_data/data/006_forg'
image_directory1 = 'signature_data/data/006'

# Plot images from the directory
plot_images_from_directory(image_directory)
print()
print()
print()
plot_images_from_directory(image_directory1)
```
